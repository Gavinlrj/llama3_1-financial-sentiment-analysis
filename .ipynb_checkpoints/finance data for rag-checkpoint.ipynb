```python
import pandas as pd

df = pd.read_csv('CBS-News.csv')
df
```

```python
import pandas as pd

# 假设你的数据已经加载到 df 中
df_2024 = df[df['date'].str.startswith('2024')]

# 查看筛选结果
df_2024
```

```python
# 选择所需的三列
df_2024_selected = df_2024[['date', 'title', 'content']]

# 保存为新的 CSV 文件
df_2024_selected.to_csv('filtered_2024_data.csv', index=False)
```

```python
import pandas as pd

# 读取已有的 2024 年数据文件
df_2024_selected = pd.read_csv('filtered_2024_data.csv')

# 选择可能与金融相关的关键词
finance_keywords = ['finance', 'market', 'stock', 'economy', 'investment', 'trading', 'crypto']

# 筛选包含金融关键词的行
df_finance = df_2024_selected[df_2024_selected['content'].str.contains('|'.join(finance_keywords), case=False, na=False)]

# 将筛选结果保存为新的 CSV 文件
df_finance.to_csv('financial_related_2024_data.csv', index=False)
```
